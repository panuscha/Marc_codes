```python
from pymarc import map_xml, MARCWriter, MARCReader
from collections import defaultdict
import pandas as pd
```

```python
years = defaultdict(int)
with open('data/ucla test/ucla_mbk_2.mrc', 'rb') as data:
    # Nacteni marcu
    reader = MARCReader(data)
    for record in reader: 
        for field in record.get_fields('008'):
            year = field.data[slice(7,11, None)]
            years[year] += 1 
            if int(year) < 2015: 
                print(record)
```
